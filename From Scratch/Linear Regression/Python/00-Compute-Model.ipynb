```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a data sets

x = np.arange(0, 200, 10)
y = np.arange(10, 210, 10)

x,y
```

```python
# Now create train and test sets

x_train = x[:16]
x_test = x[16:]

y_train = y[:16]
y_test = y[16:]
```

```python
plt.scatter(x_train, y_train, marker = 'x', c = 'r', label = "Training Data")
plt.scatter(x_test, y_test, marker = 'x', c = 'black', label = "Test Data")
plt.legend()
```

```python
# total number of sets --- m

m = len(x_train)
m
```

```python
# weights and biases

w = 1
b = 10
```

```python
# model 

def compute_model_output(x, w, b):
    
    fwb = np.zeros(m)

    for i in range(m):
        
        fwb[i] = (w * x[i]) + b

    return fwb
```

```python
# predictions on test data

y_pred = compute_model_output(x = x_train, w = w, b= b)
y_pred
```

```python
y_train
```

```python
# Now plot predictions with actual values

plt.scatter(x_train, y_train, marker = 'x', c = 'red', label = 'Training Data')
plt.scatter(x_train, y_pred, marker = 'x', c = 'black', label = 'Predictions')
plt.legend()
```

```python
# Now we make predictions on the test data

def make_predictoins(x = x_test, w = w, b = b):

    m = len(x)

    fwb = np.zeros(m)

    for i in range(m):

        fwb[i] = (w * x[i]) + b

    return fwb
```

```python
# Now we make predictions using 'make_predictions()' function

y_pred = make_predictoins()
```
