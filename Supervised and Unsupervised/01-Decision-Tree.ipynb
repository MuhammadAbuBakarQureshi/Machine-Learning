```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score

le = LabelEncoder()
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
features = ["pclass", "sex", "age", "embarked", "fare"]
target = ["survived"]
```

```python
## fill missing values for age and embarked as we are taking it as featuers and leave others

imp_median = SimpleImputer(strategy="median")
titanic[["age"]] = imp_median.fit_transform(titanic[["age"]])

imp_freq = SimpleImputer(strategy="most_frequent")
titanic[["embarked"]] = imp_freq.fit_transform(titanic[["embarked"]])
```

```python
titanic.isnull().sum()
```

```python
# encode

titanic["sex"] = le.fit_transform(titanic["sex"])
titanic["embarked"] = le.fit_transform(titanic["embarked"])
```

```python
X = titanic[features]
y = titanic[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

## Decision Tree Model - no pruning

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

```python
plt.figure(figsize=(18, 10))
plot_tree(
    model,
    feature_names=list(X.columns),
    class_names=["Died", "Survived"],
    filled=True,
    max_depth=2
)

plt.tight_layout()
plt.show()
```

## Decision Tree Model - with pre pruning

### Experiment

```python
max_depth = [2, 3, 4, 5, 6, 7, 8, 9, 10]
min_samples_split = [5, 10, 15, 20, 25, 30, 35]

previous = -10

for depth in max_depth:

    for min_split in min_samples_split:

        model = DecisionTreeClassifier(max_depth=depth, min_samples_split=min_split)

        model.fit(X_train, y_train)

        if previous < model.score(X_test, y_test):
            
            previous = model.score(X_test, y_test)
            accuracy_message = f"[INFO] Depth={depth}, Split={min_split}, Accuracy {model.score(X_test, y_test)}"

accuracy_message
```
