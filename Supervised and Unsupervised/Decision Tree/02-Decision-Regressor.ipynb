```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df = load_diabetes(as_frame=True).frame
```

```python
df.head()
```

```python
X = df.drop(columns=["target"])
y = df[["target"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
```

```python
y_train
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
    
print(f"MSE train: {mean_squared_error(y_train, y_pred_train)}")
print(f"MSE test: {mean_squared_error(y_test, y_pred_test)}")
    
print(f"r2_score train: {r2_score(y_train, y_pred_train)}")
print(f"r2_score test: {r2_score(y_test, y_pred_test)}")
```

## Decision Tree Model - with pre pruning

```python
max_depth = [2, 3, 4, 5, 6, 7, 8, 9, 10]
min_samples_split = [5, 10, 15, 20, 25, 30, 35]
min_samples_leaf = [5, 10, 15, 20, 25, 30, 35]

previous = -10

for depth in max_depth:

    for min_split in min_samples_split:

        for min_leaf in min_samples_leaf:

            model = DecisionTreeRegressor(max_depth=depth, min_samples_split=min_split, min_samples_leaf=min_leaf)

            model.fit(X_train, y_train)

            if previous < model.score(X_test, y_test):
                
                previous = model.score(X_test, y_test)
                accuracy_message = f"[INFO] Depth={depth}, Split={min_split}, Leaf={min_leaf}"
                message = f"MSE train: {mean_squared_error(y_train, y_pred_train)}\nMSE test: {mean_squared_error(y_test, y_pred_test)}\n\nr2_score train: {r2_score(y_train, y_pred_train)}\nr2_score test: {r2_score(y_test, y_pred_test)}"

print(f"{accuracy_message}\n{message}")
```

```python
plt.figure(figsize=(18, 10))

plot_tree(
    model,
    feature_names=X.columns,
    filled=True
)

plt.tight_layout()
```
