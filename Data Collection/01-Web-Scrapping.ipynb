```python
import requests
import os
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://books.toscrape.com/"
```

```python
try:
    response = requests.get(URL)
    response.raise_for_status()

except requests.exceptions.RequestException as e:

    print(f"Error: {e}")
```

```python
if response.status_code == 200:

    with open('scrapped_data/page_1.html', 'w') as file:
        
        if file.writable():

            print(f"Writing data in file....")
            file.write(response.text)
            print(f"Data stored")

else:

    print(f"Error: {response.status_code}")
```

```python
for page_number in range(1, 51):

    URL = f"https://books.toscrape.com/catalogue/page-{page_number}.html"

    print(f"\n\n[INFO] Scrapping: {URL}")

    try:

       response = requests.get(URL)
       response.raise_for_status()

    except requests.exceptions.RequestException as e:

        print(f"Error: {e}")

    SAVE_FILE_PATH = f"scrapped_data/page_{page_number}.html"

    if response.status_code == 200:

        print(f"[INFO] Saving scrapped data")

        with open(SAVE_FILE_PATH, 'w') as file:

            if file.writable():

                file.write(response.text)

            file.close()

        print(f"[INFO] Data saved at: {SAVE_FILE_PATH}")

    else:

        print(f"Error: {response.status_code}")
```

```python
books = []

for page_number in range(1, 51):

    print(f"[INFO] Reading file No :  {page_number}")

    with open(f"scrapped_data/page_{page_number}.html", 'r') as file:

        html_content = file.read()

        soup = BeautifulSoup(html_content, 'lxml')

        all_h3 = soup.find_all("h3")

        for h3 in all_h3:

            title = h3.find_next()['title']

            price = h3.find_parent("article").find_all('div')[-1].find('p').get_text()[2:]
            
            books.append({
                "title": title,
                "price": price
            })
```

```python
book_table = pd.DataFrame(books)
```

```python
book_table.to_csv('cleaned_data/books.csv', index=False)
```
