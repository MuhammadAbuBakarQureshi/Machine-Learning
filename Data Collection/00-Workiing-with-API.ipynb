```python
import requests
import pandas as pd
```

```python
URL = "https://www.freetogame.com/api/games"
```

```python
try:
    response = requests.get(URL, timeout=5)
    response.raise_for_status()
except requests.exceptions.RequestException as e:

    print(f"Error: {e}")
```

```python
data = response.json()
```

```python
data
```

```python
json_data = pd.json_normalize(data)
```

```python
filtered_json_data = json_data[['title', 'genre', 'platform', 'publisher', 'release_date', 'game_url']]
json_data
```
